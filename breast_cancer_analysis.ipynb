```python
x = df_cancer.drop(['target'], axis = 1)
```

'target' in x.columns to check if target column is still present

```python
x
```

```python
y = df_cancer['target']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test, y_train,y_test =  train_test_split(x,y,test_size = 0.2, random_state = 5)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
model = SVC()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
cm = confusion_matrix(y_test,y_pred)
```

```python
sns.heatmap(cm, annot = True)
```

```python
min_train = x_train.min()
```

```python
range_train = x_train.max()-min_train
```

```python
x_train_scaled = (x_train-min_train)/range_train
```

```python
sns.scatterplot(x = x_train['mean area'], y =x_train['mean smoothness'], hue = y_train)
```

```python
sns.scatterplot(x = x_train_scaled['mean area'], y =x_train_scaled['mean smoothness'], hue = y_train)
```

```python
model.fit(x_train_scaled,y_train)
```

```python
x_test_min = x_test.min()
range_x_test = x_test.max()-x_test_min
x_test_scaled = (x_test- x_test_min)/range_x_test
y_pred_improv = model.predict(x_test_scaled)
```

```python
cm_improv = confusion_matrix(y_test,y_pred_improv)
```

```python
sns.heatmap(cm_improv, annot = True)
```

```python
print(classification_report(y_test,y_pred_improv))
```

```python
param_grid = {'C': [0.1,1,10,100], 'gamma': [1,0.1,0.01,0.001], 'kernel': ['rbf']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(SVC(),param_grid,refit = True, verbose = 4)
```

```python
grid
```

```python
grid.fit(x_train_scaled, y_train)
```

```python
grid.best_params_
```

```python
sns.heatmap(cm_grid, annot = True)
```
